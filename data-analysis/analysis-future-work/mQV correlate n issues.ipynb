```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import matplotlib     
font = {'size'   : 12}
matplotlib.rc('font', **font)
```

```python
def to_array(string):
    return np.array([float(el) for el in string.replace("[", "").replace("]","").split(" ")])
def WL_per_issue(payoffs):
    return (payoffs < 0) * 1
```

```python
df = pd.read_csv('multi-QV-with-Polling correlate-n-issues-table.csv')
df['total-payoff-per-issue'] = df['total-payoff-per-issue'].apply(to_array)
df['mean-payoff'] = df['total-payoff-per-issue'].apply(np.mean)
df['WL-per-issue'] = df['total-payoff-per-issue'].apply(WL_per_issue)
df['mean-WL'] = df['WL-per-issue'].apply(np.mean)
display(df)
```

```python
adf = df[df['correlate-n-issues'] == 10]
ax = sns.scatterplot(x="issues-correlation", 
                     y="mean-WL", 
                     data = adf)
ax.set(ylim= (0, 1))
plt.title("QV wellfare-loss vs. issues-correlation-)
plt.show()

```
