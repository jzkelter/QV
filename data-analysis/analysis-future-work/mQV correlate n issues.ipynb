```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import matplotlib     
font = {'size'   : 12}
matplotlib.rc('font', **font)
```

```python
def to_array(string):
    return np.array([float(el) for el in string.replace("[", "").replace("]","").split(" ")])
def WL_per_issue(payoffs):
    return (payoffs < 0) * 1
```

```python
df = pd.read_csv('multi-QV-with-Polling correlate-n-issues-table.csv')
df['WL-per-issue'] = df['mean total-payoff-per-issue per run'].apply(WL_per_issue)
df['mean-WL'] = df['WL-per-issue'].apply(np.mean)
display(df)
```

```python
adf = df[df['correlate-n-issues'] == 10]
ax = sns.scatterplot(x="issues-correlation", 
                     y="mean total-payoff-per-issue per run", 
                     data = adf)
ax.set(ylim= (2, 6))
ax.set(xlim=(-1,1))
plt.title("QV wellfare-loss vs. correlate-10-issues")
plt.show()


```

Takeaways: higher correlation generates noise among the mean total payoff, and lower correlation leads to higher mean total-payoff-per-issue
