```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import matplotlib     
font = {'size'   : 12}
matplotlib.rc('font', **font)
```

```python
def color(x):
    if x == "1. prop8-mean>0":
        return 'blue'
    else:
        return 'orange'



df = pd.read_csv('QV Prop8 correlating pmp with utilities data.csv')
df['wellfare-loss'] = 1 - df['mean payoff-sign-list']
df['color'] = df['calibration'].apply(color)

df2 = pd.read_csv('QV varying-mean-utility-table 2.csv')
df2['wellfare-loss'] = 1 - df2['average payoff']


display(df)

display(df2)
```

```python
#ax = sns.scatterplot(x="utility-pmp-correlation", 
#                     y="wellfare-loss",
#                     hue="color",
#                     data = df)
#plt.title("blue='1.prop8-mean>0'  orange='2.prop8-mean=0'")

df2 = pd.read_csv('QV prop-8-correlate-pmp-utils-table.csv')
df2['wellfare-loss'] = 1 - df2['average payoff']

adf2 = df2[df2["number-of-voters"] == 1001]


ax2 = sns.scatterplot(x="utility-pmp-correlation",
                     y="wellfare-loss",
                     data = adf2)
```

Now lets confirm the differences in trends that we see above: Correlating pmp with utilities will rarely swing a vote when it's mean utility is far from zero, any "noise" that it would generate wouldn't matter because the vote would still be made in the same direction as the one that would minimize wellfare-loss.

```python
adf2 = df2[df2["majority-mean-utility"] == 0.0]
#2 options, passionate group on side of mean
#           passionate group on opposite side of mean 
#           at higher correlation, this will cause option 1 to succeed but option 2 to fail,
#           this opposite is true at negative correlation
#           While there is some correlation between the side of the passionate group and the side of the mean, it wasn't enough to create a significant difference
#           I think this is the reason that we aren't seeing much change a normal distribution, it balances out (to an extent)

ax2 = sns.scatterplot(x="utility-pmp-correlation", 
                      y="wellfare-loss",
                      data = adf2)
plt.title("wellfare-loss with majority-mean-utility = 0")
```

```python
df = pd.read_csv('QV passionate-minority-pmp-correlation-table (0.04 mean) 2.csv')
df['wellfare-loss'] = 1 - df['average payoff']
df['wellfare-loss range'] = 1 - df['mean payoff-sign-list']

display(df)

yerrormin = df['min']
yerrormax = df['max']
yerror = [yerrormin, yerrormax]


ax = sns.scatterplot(x="utility-pmp-correlation", 
                     y="wellfare-loss",
                     data = df)


plt.errorbar(df['utility-pmp-correlation'], df['wellfare-loss'], yerr=yerror, ecolor = 'orange', fmt='o')

plt.title("passionate minority, 0.04 mean, varying utility-pmp correlation")

plt.show()
```

```python
#Tipping point
#It definitely looks to be linear from close, but I will test again at higher means
df = pd.read_csv('QV passionate-minority-pmp-correlation-table (tipping).csv')
ax = sns.scatterplot(x="mean-utility", 
                     y="tipping-point",
                     data = df)
plt.title("passionate minority")
```
