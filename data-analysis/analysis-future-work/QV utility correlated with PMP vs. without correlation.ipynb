```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import matplotlib     
font = {'size'   : 12}
matplotlib.rc('font', **font)
```

```python
def to_array(string):
    return np.array([float(el) for el in string.replace("[", "").replace("]","").split(" ")])
```

```python
df = pd.read_csv('QV r-and-pmp-table.csv', error_bad_lines=False)
df['wellfare-loss'] = 1 - df['mean payoff-sign-list']

df2 = pd.read_csv('1p1v r-and-pmp-table.csv', error_bad_lines=False)
df2['wellfare-loss'] = 1 - df2['mean payoff-sign-list']
```

```python
display(df)
display(df2)
```

```python
adf = df[df["majority-mean-utility"] == 0]
ax = sns.scatterplot(x="r", 
                     y="wellfare-loss", 
                     hue = "majority-utility-stdev",
                     data = adf)
ax.set(ylim= (0, 0.1),
       xlabel = "correlation")
plt.title("QV wellfare-loss vs. correlation; $\mu=0$")
plt.show()

adf2 = df2[df2["majority-mean-utility"] == 0]
ax = sns.scatterplot(x="r", 
                     y="wellfare-loss", 
                     hue = "majority-utility-stdev",
                     data = adf2)
ax.set(ylim= (0, 0.6),
       xlabel = "correlation")
plt.title("1p1v wellfare-loss vs. correlation; $\mu=0$")
plt.show()
    
```
